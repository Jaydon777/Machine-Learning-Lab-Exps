## Linear Regression

```python
import pandas as pd
```

```python
df = pd.read_csv('Car_sales.csv')
df
```

```python
df.isnull().sum()
```

## X - independent 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity', 'Fuel_efficiency', 'Power_perf_factor'
## Y - dependent __year_resale_value

```python
columns_of_interest = ['Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 
                       'Fuel_capacity', 'Fuel_efficiency', 'Power_perf_factor', '__year_resale_value']
df_cleaned = df.dropna(subset=columns_of_interest)

X = df_cleaned[['Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 
                'Fuel_capacity', 'Fuel_efficiency', 'Power_perf_factor']]
Y = df_cleaned['__year_resale_value']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_normalized,Y,test_size = 0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
c = lr.intercept_
```

```python
c
```

```python
m = lr.coef_
m
```

```python
y_pred_train = lr.predict(X_train)
```

```python
y_pred_train
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_train, y_pred_train)
```
